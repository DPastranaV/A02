**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
sp.var("x")
def f(x) :
    return (1260 + 117*x - 9*x**2)

f(x)
```

```python
df = sp.diff(f(x), x)
df
```

```python
pc = sp.solve(df, x)
pc
```

```python
xn = np.linspace(4, 9, 1000)
yn = f(xn)

plt.figure()
plt.plot(xn, yn)
plt.grid()

plt.plot(pc[0], f(pc[0]), "*r")
```

```python
ddf = sp.diff(df, x)
ddf
```

```python
sp.diff(f(x), x, 2)
```

Por tanto, por el criterio de la segunda derivada y después de graficar, el punto crítico en $13/2$ es un máximo global.

$f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
sp.var("x")
def f(x) :
    return 1/4*x**4 + 1/3*x**3 - x**2

f(x)
```

```python
df = sp.diff(f(x), x)
df
```

```python
pc = sp.solve(df, x)
pc
```

```python
xn = np.linspace(-3, 1.7, 1000)
yn = f(xn)

plt.figure()
plt.plot(xn, yn)
plt.grid()
plt.plot(pc[0], f(pc[0]), "*r")
plt.plot(pc[1], f(pc[1]), "*r")
plt.plot(pc[2], f(pc[2]), "*r")
```

Por tanto, después de graficar, el punto crítico en $-2$ es un mínimo global, el punto crítico en $0$ es un máximo local, y el punto crítico en $1$ es un mínimo local.

El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

```python
import sympy as sp
import numpy as np

sp.var("x")
sp.var("r")
sp.var("h")

perimetro = 2*x + r - 10
perimetro
```

```python
x_sol = sp.solve(perimetro, x)[0]
x_sol
```

```python
altura = sp.sqrt(x**2 - (r/2)**2)
h_sol = altura.subs(x, x_sol)
h_sol
```

```python
volumen = (np.pi * r**2 * h)/3
volumen
```

```python
volumen_r = volumen.subs(x, x_sol).subs(h, h_sol)
volumen_r
```

```python
dvolumen_r = sp.diff(volumen_r, r)
dvolumen_r
```

```python
r_num = sp.solve(dvolumen_r, r)
r_num
```

```python
x_num = x_sol.subs(r, r_num[0])
x_num
```

```python
2*x_num + r_num[0]
```

```python
float(sp.diff(dvolumen_r, r).subs(r, r_num[0]))
```

```python
x_num = x_sol.subs(r, r_num[1])
x_num
```

```python
2*x_num + r_num[1]
```

```python
float(sp.diff(dvolumen_r, r).subs(r, r_num[1]))
```

Los lados deben de medir  $3cm$  y la base  $4cm$  para que el volumen sea el mayor posible



Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

```python
import sympy as sp
import numpy as np

sp.var("x")
sp.var("y")

perimetro = 2*x + y - 6
perimetro
```

```python
x_sol = sp.solve(perimetro, x)[0]
x_sol
```

```python
area = x*y
sustituir_x = area.subs(x, x_sol)
sustituir_x
```

```python
dsustituir_x = sp.diff(sustituir_x, y)
dsustituir_x
```

```python
y_num = sp.solve(dsustituir_x, y)
y_num
```

```python
x_num = x_sol.subs(y, y_num[0])
x_num
```

```python
2*x_num + y_num[0]
```

```python
float(sp.diff(dsustituir_x, y).subs(y, y_num[0]))
```

Las medidas de los postes deben de ser de  $1.5m$  y la medida del travesaño debe de ser de $3m$ para que el área sea máxima
