$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
sp.var("x")
def f(x) :
    return (1260 + 117*x - 9*x**2)

f(x)
```

```python
df = sp.diff(f(x), x)
df
```

```python
pc = sp.solve(df, x)
pc
```

```python
xn = np.linspace(4, 9, 1000)
yn = f(xn)

plt.figure()
plt.plot(xn, yn)
plt.grid()

plt.plot(pc[0], f(pc[0]), "*r")
```

```python
ddf = sp.diff(df, x)
ddf
```

```python
sp.diff(f(x), x, 2)
```

```python
def df(x) :
    return -18*x + 117

def ddf(x) :
    return -18*x/x

xn = np.linspace(-10, 10, 1000)
yn = f(xn)
dyn = df(xn)
ddyn = ddf(xn)

plt.figure()
plt.plot(xn, yn)
plt.plot(xn, dyn)
plt.plot(xn, ddyn)
plt.grid()

```

Por tanto, por el criterio de la segunda derivada, $f(-18)=0$ es un máximo global.

$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
sp.var("x")
def f(x) :
    return (5 + x + 4/x)

f(x)
```

```python
df = sp.diff(f(x), x)
df
```

```python
pc = sp.solve(df, x)
pc
```

```python
xn = np.linspace(-5, 5, 1000)
yn = f(xn)

plt.figure()
plt.plot(xn, yn)
plt.grid()

plt.plot(pc[0], f(pc[0]), "*r")
plt.plot(pc[1], f(pc[1]), "*r")
```

```python
ddf = sp.diff(df, x)
ddf
```

```python
sp.diff(f(x), x, 2)
```

```python
def df(x) :
    return 1 - 4/x**2

def ddf(x) :
    return 8/x**3

xn = np.linspace(1, 10, 1000)
yn = f(xn)
dyn = df(xn)
ddyn = ddf(xn)

plt.figure()
plt.plot(xn, yn)
plt.plot(xn, dyn)
plt.plot(xn, ddyn)
plt.grid()
```
